```python
#we can see that still we have 2 columns namely 'desc' and 'mths_since_last_delinq' which have more null values and
#these are loan behavior column whose data we will get after the loan is approved. So we dont need them at the loan applicant time.
#So lets delete them too.
data = data.drop(['desc', 'mths_since_last_delinq'], axis=1)
```

```python
#now lets check the rows whcih have null values
data.isnull().sum(axis=1)
```

```python
# we couldnt find any row which have more than 5 missing values
```

```python
#now, after removing the null data, lets summarize the available data
#for analysis, we need to convert the data into proper format
data.info()
```

```python
#here 1 of the column 'int_rate' is in object datatype which should be float. so lets convert it
data['int_rate']
```

```python
#lets seperate the numeric part and '%' symbol to convert into float datatype
if (data['int_rate'].dtype != 'float64'):
    data['int_rate'] = data['int_rate'].apply(lambda x : pd.to_numeric(x.split('%')[0]))
```

```python
data.info()
```

```python
#Similarly, we also have 1 more column named 'emp_length' which is of Object type which need to be converted into float.
data['emp_length']
```

```python
data = data[~data['emp_length'].isnull()]
```

```python
import re
#loan['emp_length'] = loan['emp_length'].apply(lambda x: re.findall('\d+', str(x))[0])
data['emp_length'] = data['emp_length'].apply(lambda x: re.findall('\d+', str(x))[0])
```

```python
#lets convert this to numeric
data['emp_length'] = data['emp_length'].apply(lambda x: pd.to_numeric(x))
data['emp_length']
```

```python
#Now, we should the type of the columns whether they are DEMOGRAPHIC VARIABLES, LOAN CHARACTERISTICS VARIABLES or BEHAVIOURAL VARIABLES.
#To decide that whether we should approve the loan applicants or not, we basically only need loan characteristics variables initally.
#So lets discard the rest of the columns from our dataset.
```

```python
#lets first delete behavioural variables
behavioural_vars = ["delinq_2yrs", "earliest_cr_line", "inq_last_6mths", "open_acc", "pub_rec", "revol_bal", "revol_util",
  "total_acc", "out_prncp", "out_prncp_inv", "total_pymnt", "total_pymnt_inv", "total_rec_prncp", "total_rec_int", "total_rec_late_fee",
  "recoveries", "collection_recovery_fee", "last_pymnt_d", "last_pymnt_amnt", "last_credit_pull_d", "application_type"]
data = data.drop(behavioural_vars, axis=1)
```

```python
#lets discard demographic variables also
data = data.drop(['title', 'url', 'zip_code', 'addr_state'], axis=1)
```

```python
#Now, we are left with only those columns which will help us decide to approve/reject loan applicant
```

```python
data['loan_status']
```

```python
#the main column here available now known as target variable is loan_status. But we should first clean the loan_status values.
#As we can see, loan_status have 3 values Fully Paid, Current, Charged Off. Right now, we dont need Current details as they are already running. 
#So lets count the loan applicants which dont have status as current.
data = data[data['loan_status'] != 'Current']
```

```python
#For other 2 values of loan_status, lets convert them into integer for easy manipulation. 
#So giving value=0 for Fully Paid and value=1 for Charged Off and converting them to integer
data['loan_status'] = data['loan_status'].apply(lambda x: 0 if x=='Fully Paid' else 1)
data['loan_status'] = data['loan_status'].apply(lambda x: pd.to_numeric(x))
```

```python
data['loan_status'].value_counts()
```

```python
######### UNIVARIATE ANALYSIS #########
```

```python
#As we will analyse different variables against loan_status, and plots different graphs, so instead of writing same lines of code everytime,
#lets make a function tomplot a graph
```

```python
# lets define a function to plot loan_status across categorical variables
def plot_cat(var):
    plt.xlabel(var)
    plt.ylabel('Loan Status')
    plt.title(f'Loan Status vs {var}')
    sns.barplot(x=var, y='loan_status', data=data, hue=var, palette='autumn_r')    
    plt.show()
```

```python
#lets plot the graph of laon_status vs grade to understand that which grade have how much default value
plot_cat('grade')
```

```python
#From the above plot, we can conclude that as the grade goes from A to G, the default value increases. 
#Means loan with grade A is less risky than loan with grade G
```

```python
#Now lets analyse the riskyness against the term period.
plot_cat('term')
```

```python
#Here we can see that the loan with 60 months of period have more defaulters than the loan with 36months of period. 
```

```python
#Lets analyse against the sub-grades. By default we know that as the sub-grade increases from grade A1 to G5, risk also increases.
#But lets check that whether its true for out dataset or not
plot_cat('sub_grade')
```

```python
#As the number of ticks are more in x axis here, lets increase the plot size.
plt.figure(figsize=(20,10))
plot_cat('sub_grade')
```

```python
#Lets analyse against home ownership
plot_cat('home_ownership')
```

```python
#We can see that defaulter rate against home ownership is almost equal for those who have rented, owned or mortgage their home.
```

```python
#Lets plot against purpose of loan 
plot_cat('purpose')
```

```python
#increase the figure size so that we can easily watch the purposes
plt.figure(figsize=(15,10))
plot_cat('purpose')
```

```python
#here the ones who have taken loan for their small business have defaulted most
```

```python
#Now, lets check the loan defaulters month and year wise. For that we need to revise the data column values first.
data['issue_d']
```

```python
#We need to extract the year and month from the issue date column values.SO for that we will be using datetime strptime function
from datetime import datetime
data['issue_d'] = data['issue_d'].apply(lambda x: datetime.strptime(x, '%b-%y'))
#making new column for issue year
data['year'] = data['issue_d'].apply(lambda x: x.year)
#making new column for issue month
data['month'] = data['issue_d'].apply(lambda x: x.month) 
```

```python
#So now we as we have extracted month and year from the issue date, lets anylse loan status against loan issue year
plot_cat('year')
```

```python
#we can see that in the year 2007, most defaulters have been noted and then gradually the defaulters rate is decreasing 
```

```python
#anaylse against loan issue month
plot_cat('month')
```

```python
#In the month of December, people have defaulted the most
```

```python
################### DEFAULTER'S RATE FOR CONTINOUS VARIABLES ###################
```

```python
#loan Amount is the continous variable. So lets anaylse it.
data['loan_amnt']
```

```python
#This is the numerical continous amount 
sns.displot(data['loan_amnt'])
```

```python
#Not lets divide this numerical amount into category to make bins.
#So we will bin the loan amount into low, medium, high, very high category
def check_loan_amnt(x):
    if (x < 5000):
        return 'low'
    elif (x >= 5000 and x<15000):
        return 'medium'
    elif (x>=15000 and x<25000):
        return 'high'
    else:
        return 'very high'

data['loan_amnt_bin'] = data['loan_amnt'].apply(lambda x: check_loan_amnt(x))
plot_cat('loan_amnt_bin')
```

```python
#Here we anylsed that the loan with more than 25000 loan amount have high defaulters 
#while the loan with amount between 5000 and 15000 has very less defaulters.
```

```python
#Similarly lets check the funded amount by investors for loan applicants
data['funded_amnt_inv']
```

```python
#lets bin them into the same categories as loan amount
data['funded_inv_bin'] = data['funded_amnt_inv'].apply(lambda x: check_loan_amnt(x))
plot_cat('funded_inv_bin')
```

```python
#Now analyse against funded amount. So for that lets bin the funded amount variable into low, medium, high
def funded_amount(x):
    if x <= 5000:
        return 'low'
    elif x > 5000 and x <= 15000:
        return 'medium'
    else:
        return 'high'
    
data['funded_amnt_bin'] = data['funded_amnt'].apply(lambda x: funded_amount(x))
plot_cat('funded_amnt_bin')
```

```python
#We analysed that the loan with high funded amount have more defaulters
```

```python
#Now we will analyse the interest rate values.
#So first bin them into 3 categories - low, medium and high
def int_rate(n):
    if n <= 10:
        return 'low'
    elif n > 10 and n <=15:
        return 'medium'
    else:
        return 'high'

data['int_rate_bin'] = data['int_rate'].apply(lambda x: int_rate(x))
plot_cat('int_rate_bin')
```

```python
#Here we can see that, the loan with high interest rate have more defaulters
```

```python
#Now analysing the dti variable
def dti(n):
    if n <= 10:
        return 'low'
    elif n > 10 and n <=20:
        return 'medium'
    else:
        return 'high'
    

data['dti_bin'] = data['dti'].apply(lambda x: dti(x))
plot_cat('dti_bin')
```

```python
#Loan with higher dti have higher defaulters
```

```python
# anaylse against installment for loan. So bin them into low, medium, high category and plot the graph
def installment(n):
    if n <= 200:
        return 'low'
    elif n > 200 and n <=400:
        return 'medium'
    elif n > 400 and n <=600:
        return 'high'
    else:
        return 'very high'
    
data['installment_bin'] = data['installment'].apply(lambda x: installment(x))
plot_cat('installment_bin')
```

```python
#we can see that the higher the installment, the higher the defaulters.
```

```python
#Lets anaylse against annual income of the loan applicant. So for that bin them into low, medium, high and very high.
def annual_income(n):
    if n <= 50000:
        return 'low'
    elif n > 50000 and n <=100000:
        return 'medium'
    elif n > 100000 and n <=150000:
        return 'high'
    else:
        return 'very high'

data['annual_inc_bin'] = data['annual_inc'].apply(lambda x: annual_income(x))
plot_cat('annual_inc_bin')
```

```python
#Here, the one who have low income have defaulted more.
```

```python
#Now, lets anaylse against employment length
data['emp_length']
```

```python
#So bin them into categories - fresher, junior, senior and expert according to their work experience
def emp_length(n):
    if n <= 1:
        return 'fresher'
    elif n > 1 and n <=3:
        return 'junior'
    elif n > 3 and n <=7:
        return 'senior'
    else:
        return 'expert'

data['emp_length_bin'] = data['emp_length'].apply(lambda x: emp_length(x))
plot_cat('emp_length_bin')
```

```python
#Here junior are bit less defaulters compared to expert and seniors.
```
